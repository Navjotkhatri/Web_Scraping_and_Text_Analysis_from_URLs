Polarity Score: This is the score that determines if a given text is positive or negative in nature.

It is calculated by using the formula:


Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)


```python
df_urls_text['Polarity_Score'] = (df_urls_text['Positive_Score']-df_urls_text['Negative_Score'])/((df_urls_text['Positive_Score']+df_urls_text['Negative_Score'])+ 0.000001)
```

One common definition is that a complex word is a word with three or more syllables. So I used this ..

Percentage of Complex words = the number of complex words / the number of words


```python
from nltk.corpus import cmudict
```

```python
import nltk
nltk.download('cmudict')
```

```python
d = cmudict.dict()

def count_syllables(word):
    if word.lower() in d:
        return [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]][0]
    else:
        return 0

def is_complex_word(word):
    return count_syllables(word) >= 3

def complex_words(text):
  return [word for word in text if is_complex_word(word)]
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
def remove_punctuation(word):
    return word.translate(str.maketrans('', '', string.punctuation))

# Function to remove stop words and punctuation
def clean_text(text):
    cleaned_words = [remove_punctuation(word) for word in text if word.lower() not in stop_words and remove_punctuation(word) != '']

    return cleaned_words
```

```python
df_urls_text['Cleaned_Tokens2'] = df_urls_text['tokens'].apply(clean_text)
```

```python
df_urls_text['Word_Count'] = df_urls_text['Cleaned_Tokens2'].apply(len)
```

```python
def count_syllables(word):
    word = word.lower()
    vowels = "aeiou"
    syllable_count = 0

    if len(word) <= 3:
        return 1

    if word[-2:] == "es" or word[-2:] == "ed":
        word = word[:-2]

    for char in word:
        if char in vowels:
            syllable_count += 1

    if word.endswith("e"):
        if syllable_count > 1:
            syllable_count -= 1

    return syllable_count if syllable_count > 0 else 1

def sys_count(text):
  return [count_syllables(word) for word in text]
```

```python
df_urls_text['Syllable_Count_Per_Word'] = df_urls_text['tokens'].apply(sys_count)
```

```python
import nltk
import re
import string
```

```python
def count_personal_pronouns(tokens):
    pronouns = ['I', 'we', 'my', 'ours', 'us']
    pronoun_pattern = re.compile(r'\b(?:' + '|'.join(pronouns) + r')\b', re.IGNORECASE)
    count = sum(1 for token in tokens if pronoun_pattern.match(token) and token != 'US')
    return count
```

```python
df_urls_text['Personal_Pronouns'] = df_urls_text['tokens'].apply(count_personal_pronouns)
```

Average Word Length is calculated by the formula:
Sum of the total number of characters in each word/Total number of words


```python
def calculate_average_word_length(tokens):
    total_chars = sum(len(word) for word in tokens)
    total_words = len(tokens)
    avg_word_length = total_chars / total_words if total_words > 0 else 0
    return avg_word_length
```

```python
df_urls_text['Average_Word_Length'] = df_urls_text['tokens'].apply(calculate_average_word_length)
```
